```python
# install dependencies
# pip intall plotly
# pip install pandas

import pandas as pd
import plotly.express as px
```

```python
data = [
    ("people", 0.020635),
    ("don", 0.018773),
    ("trump", 0.017617),
    ("just", 0.016816),
    ("like", 0.016715),
    ("think", 0.016471),
    ("gt", 0.014162),
    ("right", 0.010808),
    ("vote", 0.009826),
    ("really", 0.009123),
    ("party", 0.008964),
    ("going", 0.008902),
    ("way", 0.008826),
    ("know", 0.008755),
    ("want", 0.008643),
    ("make", 0.008461),
    ("good", 0.008400),
    ("comments", 0.008370),
    ("doesn", 0.008353),
    ("government", 0.008268),
]

data = pd.DataFrame(data, columns=["term", "tfidf"]).sort_values("tfidf", ascending=True)

fig = px.bar(
    data,
    x="tfidf", y="term",
    orientation="h",
    color="tfidf",
    color_continuous_scale="Viridis",
    text=data["tfidf"].map(lambda v: f"{v:.4f}")
)

fig.update_traces(textposition="outside", cliponaxis=False)
fig.update_layout(
    title=dict(text="top 20 words by TF-IDF score in r/politicaldiscussion", x=0.5, xanchor="center"),
    xaxis_title="TF-IDF Score",
    yaxis_title="",
    margin=dict(l=20, r=20, t=60, b=20),
    template="plotly_white",
)

fig.update_layout(
    coloraxis_colorbar=dict(
        title="TF-IDF",
        x=1.02,     # horizontal position (default = 1.0)
        y=0.5,      # vertical center (0=bottom, 1=top)
        len=0.75,   # shrink length (0â€“1)
        thickness=15  # width in px
    )
)

fig.write_html("tfidf_top20_interactive.html", include_plotlyjs="inline")


fig.show()
```
